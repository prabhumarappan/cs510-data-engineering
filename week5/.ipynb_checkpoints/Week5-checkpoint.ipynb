```python
import re
import math

def fix_date_issues(d):
    if type(d) == float and math.isnan(d):
        return pd.NA
    d = re.sub('\s?(\[[\d]+\]|\[[\d]+[^\d]+\]|-\d+|, \d+|[^\d]+)', '', d)
    if len(d) > 4:
        return int(d[:4])
    if len(d) == 0:
        return pd.NA
    if len(d) < 4:
        return int("%04d" % int(d))
    
    return int(d)

for index, row in df['Date of Publication'].iteritems():
    df['Date of Publication'][index] = fix_date_issues(row)

df['Date of Publication'] = pd.to_numeric(df['Date of Publication'])
```
